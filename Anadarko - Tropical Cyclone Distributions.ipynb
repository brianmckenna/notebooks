Iterate through all the known tracks in Southern Indian Ocean.

If any point along track intersects the buffer defined by radius from point, include.

```python
def haversine(lat1, lon1, lat2, lon2):
 
  R = 6372.8 # Earth radius in kilometers
 
  dLat = math.radians(lat2 - lat1)
  dLon = math.radians(lon2 - lon1)
  lat1 = math.radians(lat1)
  lat2 = math.radians(lat2)
 
  a = math.sin(dLat/2)**2 + math.cos(lat1)*math.cos(lat2)*math.sin(dLon/2)**2
  c = 2*math.asin(math.sqrt(a))
 
  return R * c
```

```python
points = []
tracks = []
wnds = []
fwds = []
prs_list = []
with open('tracks.json', 'r') as f:
    storms = json.load(f)
    for s in storms.keys():
        if all(key in storms[s] for key in ['lat','lon','pressure','wind','time']):
            # check length of each component is the same
            if len(set([len(x) for x in storms[s].values()])) != 1:
                print 'unequal lengths of keyed arrays, eg. len(lat) != len(lon)'
                continue
            time = storms[s]['time']
            lat = storms[s]['lat']
            lon = storms[s]['lon']
            prs = storms[s]['pressure']
            wnd = storms[s]['wind']
            

            fwd = [0.0] # filler value so all arrays same len
            # compute forward speed from lonlat and time (result will be one smaller than above arrays)
            for x in xrange(1, len(time)):
                delta_t = (np.datetime64(time[x]) - np.datetime64(time[x-1])).item().total_seconds()
                d = haversine(lat[x-1],lon[x-1],lat[x],lon[x])
                #d = 300 * d * math.cos(math.radians((lat[x-1]+lat[x])/2.0)) #TODO this for haversine?               
                fwd.append((d/delta_t)*1943.84449) #km/s to knots
            
            if len(lon) > 2 and len(lat) > 2: # need at least one point with
                # skip junk data
                if all([w == 0.0 for w in wnd]):
                    continue
                if all([p == 0.0 for p in prs]):
                    continue
                track = LineString(zip(lon,lat))

                max_wind = max(wnd)
                if max_wind < 83:
                    continue

                # store all the winds, the pressure and forward speeds here to do distribution in next cell
                wnds.append(wnd)
                prs_list.append(prs)
                fwds.append(fwd)
                # only consider if intersects the buffer AND max wind from storm > category 3 (>=96 knots)
                if track.intersects(CM9_buffer):
                    points.append((lon,lat))
                    tracks.append(track)
```

```python
pltscale = 4
npts = 10000
plt_idx = 4
```

```python
fwd_bag = [f for fl in fwds for f in fl]
fmean = np.mean(fwd_bag); fstd = np.std(fwd_bag)
fx = np.linspace(fmean-pltscale*fstd,fmean+pltscale*fstd,npts)
fy = norm.pdf(fx, loc = fmean, scale = fstd)
fx_weibull = np.linspace(-5,100,npts)
fwd_weibull_params = exponweib.fit(fwd_bag)
fy_weibull = exponweib.pdf(fx_weibull, *fwd_weibull_params)

plt.figure()
plt.plot(fx,fy,'b-',linewidth=2, label='Gaussian')
plt.title('Forward Speed')
ax = plt.gca()
ax.hist(fwd_bag, normed=True, histtype='stepfilled', alpha=0.2, label='Empirical')
plt.plot(fx_weibull, fy_weibull,'b--', linewidth=2, label='Weibull')

leg = ax.legend(loc='best',fancybox=True)
leg.get_frame().set_alpha(0.5)

plt.xlabel('Forward Speed (Knots)')
plt.ylabel('Probability Density')
plt.savefig('{0:03}-fwd-gauss-pdf.png'.format(plt_idx))
plt_idx+=1
```

```python
wnds_bag = [w for wl in wnds for w in wl if w > 0]
wmean = np.mean(wnds_bag); wstd = np.std(wnds_bag)
wx = np.linspace(wmean-pltscale*wstd, wmean+pltscale*wstd, npts)
wy = norm.pdf(wx, loc = wmean, scale = wstd)

wnds_weibull_params = exponweib.fit(wnds_bag)
wx_weibull = np.linspace(0, 200,npts)
wy_weibull = exponweib.pdf(wx_weibull, *wnds_weibull_params)

plt.figure()
plt.plot(wx, wy, 'r-', linewidth=2,label='MLE Gaussian')
plt.plot(wx_weibull, wy_weibull, 'r--', linewidth = 2, label = 'MLE Weibull')
ax = plt.gca()
ax.hist(wnds_bag, normed=True, histtype='stepfilled',alpha=0.2, label='Empirical')
plt.title('Wind Speed Distribution')
plt.xlabel('Wind Speed (Knots)')
plt.ylabel('Probability Density')
leg = ax.legend(loc='best',fancybox=True)
leg.get_frame().set_alpha(0.5)
plt.savefig('{0:03}-wnd-gauss-pdf.png'.format(plt_idx))
plt_idx+=1
```

```python
prs_bag = [p for pl in prs_list for p in pl if p > 0]
pmean = np.mean(prs_bag); pstd = np.std(prs_bag)
px = np.linspace(pmean-pltscale*pstd, pmean+pltscale*pstd, npts)
py = norm.pdf(px, loc = pmean, scale = pstd)
plt.figure()
plt.plot(px, py,'g-',linewidth=2, label='Gaussian')
ax = plt.gca()
ax.hist(prs_bag, normed=True, histtype='stepfilled', alpha=0.2, label='Empirical')
leg = ax.legend(loc='best', fancybox=True)
leg.get_frame().set_alpha(0.5)
plt.title('Pressure Distribution')
plt.xlabel('Pressure (MB)')
plt.ylabel('Probability Density')
plt.savefig('{0:03}-prs-gauss-pdf.png'.format(plt_idx))
plt_idx+=1
```
